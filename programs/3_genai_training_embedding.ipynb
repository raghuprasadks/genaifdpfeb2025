```python
from openai import OpenAI
```

```python
# Set your OpenAI API key in google colab secrets
from google.colab import userdata
OPENAI_API_KEY = userdata.get("OPENAI_API_KEY")
```

```python
client=OpenAI(api_key=OPENAI_API_KEY)
```

```python
response = client.embeddings.create(
    model="text-embedding-3-large",
    input="India is my country. I am an Indian"
)

```

```python
embedding = response.data[0].embedding
```

```python
print(embedding)
```

```python
import cohere
from google.colab import userdata
COHERE_API_KEY = userdata.get("COHERE_API_KEY")

co = cohere.Client(COHERE_API_KEY)  # Replace with your API key
```

```python
text = "India is my country. I am an Indian"
response = co.embed(texts=[text], model="large")
embedding = response.embeddings[0]
```

```python
print(embedding)
```
